## ACCURATE VISUALIZATION

### Survival rate by  gender

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()
```

This graph clearly compares survival outcomes by gender using clear labels and and prefect scale, showing that 233 of 314 female passengers (74.2%) survived, while only 109 of 577 male passengers (18.9%) survived. It accurately highlights that female passengers were nearly four times more likely to survive than male passengers, making the survival pattern easy to understand.

## Survival by Passenger Class

```python
sns.barplot(x="Pclass", y="Survived", data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

This graph accurately compares survival rates across passenger classes using a consistent scale and clear labels, showing that 1st class had the highest survival rate (63.0%), followed by 2nd class (47.3%), while 3rd class had the lowest survival rate (24.2%). It clearly demonstrates that passengers in higher classes were significantly more likely to survive than those traveling in lower classe

# Misleading visualizations:

## Port Embarkment by turnicated Y-axis


```python

plt.figure()
sns.barplot(x="embark_town", y="survived", data=df)
plt.title("Survival Rate by Port of Embarkation")
plt.xlabel("Embarkation Port")
plt.ylabel("Survival Rate")
plt.ylim(0.30, 0.60)
plt.show()
```

This misleading visual uses a truncated y-axis (0.30–0.60), which exaggerates the differences in survival rates among embarkation ports, making the gaps appear much larger than they actually are. Although the actual survival rates are approximately 33.7% for Southampton, 55.3% for Cherbourg, and 38.9% for Queenstown, the shortened axis can mislead viewers into overestimating these differences.

## Cherrypicking the data

```python
filtered = df[df["embark_town"].isin(["Cherbourg", "Southampton"])]
plt.figure()
sns.barplot(x="embark_town", y="survived", data=filtered)
plt.title("Survival Rate by Selected Ports")
plt.ylim(0,1)
plt.show()
```

This visualization is misleading because it cherry-picks only Southampton and Cherbourg, excluding Queenstown (38.9% survival rate), which presents an incomplete picture of the data. It only shows Southampton (33.7% survival) and Cherbourg (55.3% survival), making the difference appear more significant and potentially leading viewers to incorrect conclusions about survival rates across all embarkation ports.
